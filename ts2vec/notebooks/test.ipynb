```python
out = x.clone()

p = 0
reprs = []
while (1 << p) + 1 < out.size(1):
    t_out = F.max_pool1d(
        out.transpose(1, 2),
        kernel_size = (1 << (p + 1)) + 1,
        stride = 1,
        padding = 1 << p
    ).transpose(1, 2)
    reprs.append(t_out)

    print(p)
    print('kernel size:', (1 << (p + 1)) + 1)
    print('padding:', 1 << p)
    print('t_out:', t_out.shape)

    p += 1
out = torch.cat(reprs, dim=-1)
```
